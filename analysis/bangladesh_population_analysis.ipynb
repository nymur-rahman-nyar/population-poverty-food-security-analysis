```python
import pandas as pd

df = pd.read_csv("../data/raw/WPP2024_Demographic_Indicators_Medium.csv")
list(df.columns)
```

```python
# filtering the data so we have only data from Bangladesh . 
bd = df[df["Location"] == "Bangladesh"].copy()

bd.head()
```

```python
# filtered out the required columns 
# as i will be working with population related data only 
keep_cols = [
    "Location", "Variant", "Time",
    "TPopulation1July", "PopDensity", "MedianAgePop",
    "PopGrowthRate", "TFR", "NatChange", "NetMigrations"
]

bd_pop = bd[keep_cols].copy()

bd_pop.tail()

```

```python
# removing projection from the data 
bd_pop_clean = bd_pop[ (bd_pop["Time"] >= 2000) & (bd_pop["Time"] <= 2024) ].copy()

bd_pop_clean.shape
```

```python
# saving the filtered data to the processed folder 
bd_pop_clean.to_csv(
    "../data/processed/bangladesh_population_clean.csv",
    index=False
)
```
